```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("Zodiac_Analysis.db")
industrydf = pd.read_sql_query("select distinct industry from pantheon_people_import;", con)
print(industrydf)
con.close()
```

```python
#fix capitalization
#industrydf['industry'] = industrydf['industry'].str.title()
#add Industry ID to each 
industrydf['Industry_ID'] = range(1, 28 , 1)
#Reverse column order and make domain_id the index
industrydf.iloc[:,[1,0]]
industrydf = industrydf.set_index('Industry_ID')
industrydf.head(30)
```

```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("Zodiac_Analysis.db")
occdf = pd.read_sql_query("select distinct occupation from pantheon_people_import;", con)

# Verify that result of SQL query is stored in the dataframe
print(occdf)

con.close()
```

```python
#fix capitalization
#occdf['occupation'] = occdf['occupation'].str.title()
#add Industry ID to each 
occdf['Occupation_ID'] = range(1, 89 , 1)
#Reverse column order and make domain_id the index
occdf.iloc[:,[1,0]]
occdf = occdf.set_index('Occupation_ID')
occdf.head(30)
```

```python
con = sqlite3.connect("Zodiac_Analysis.db")
areadf = pd.read_sql_query("select distinct area from pantheon_people_import;", con)
con.close()
#fix capitalization
#areadf['Area'] = areadf['Area'].str.title()
#add Area ID to each 
areadf['Area_Id'] = range(1, 9 , 1)
#Reverse column order and make domain_id the index
areadf.iloc[:,[1,0]]
areadf = areadf.set_index('Area_Id')
areadf.head(10)
```

```python
#write to table in SQLlite db
connection = sqlite3.connect('Zodiac_Analysis.db')
industrydf.to_sql('Industry', connection, if_exists='replace')
occdf.to_sql('Occupation', connection, if_exists='replace')
areadf.to_sql('Area', connection, if_exists='replace')
```
