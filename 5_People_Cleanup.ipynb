```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
query=""" select ppi.name as Name, a.Area, i.Industry, o.Occupation ,z.zodiac as Zodiac_Sign
    from famous_people_import fpi
    inner join pantheon_people_import ppi on fpi.People_Lookup=ppi.People_Lookup
    inner join area a on a.area=ppi.Area
    inner join industry i on i.industry=ppi.industry
    inner join occupation o on o.occupation=ppi.occupation
    inner join Zodiac z on z.date_id=fpi.dateid"""
con = sqlite3.connect("Zodiac_Analysis.db")
df = pd.read_sql_query(query, con)
con.close()
df.head(25)
```

```python
df.to_csv('Famous_Signs.csv', index=False)
```
