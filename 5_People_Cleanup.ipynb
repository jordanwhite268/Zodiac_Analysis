```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
query='select fpi.FirstName as Fam_First_Name, fpi.LastName as Fam_Last_Name, ppi.firstname as Panth_First_Name, ppi.middlename as Panth_Middle_Name, ppi.lastname as Panth_Last_Name, a.Area_Id, i.Industry_ID, o.Occupation_ID, fpi.DateID from famous_people_import fpi left join pantheon_people_import ppi on fpi.People_Lookup=ppi.People_Lookup left join area a on a.area=ppi.Area left join industry i on i.industry=ppi.industry left join occupation o on o.occupation=ppi.occupation where Panth_First_Name is NOT NULL;'
con = sqlite3.connect("Zodiac_Analysis.db")
df = pd.read_sql_query(query, con)
con.close()
df.head(25)
```

```python
# Create Person ID and set as index.
df['Person_ID'] = range(1, (len(df)+1), 1)
df = df.set_index('Person_ID')
df.head(50)
```
