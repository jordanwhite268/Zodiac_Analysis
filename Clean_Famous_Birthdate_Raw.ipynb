```python
import pandas as pd
import numpy 

#Read in raw data file into pandas df

data = pd.read_csv(r'C:\Users\Jordan\Documents\CodeKY\Capstone_Project_Repo\Data\famous-birthdates.txt', delimiter = " ")
data.head(100)


```

```python
#drop rows without a birthmonth
data.dropna(subset = ['birthMonth'], inplace=True)
data.head(100)
```

```python
#Change number type for birthmonth and birthday to int
data_2 = data.astype({"birthMonth":'int64', "birthDay":'int64'})
data_2.head(100)
```

```python
#Rename Columns
data_2.rename(columns={'name': 'LastFirstName', 'lastname': 'Last_Name','firstname': 'First_Name',
                      'birthDate': 'Full_Birth_Date', 'birthMonth': 'Month', 'birthDay': 'Day', 'zodiac': 'Zodiac_Sign'
                      }, inplace=True)

data_2.head()
```

```python
#write dataframe to previously created database table Birthdate_Date_Import

import sqlite3
connection = sqlite3.connect('Zodiac_Analysis.db')
data_2.to_sql('birthdate_data_import', connection, if_exists='replace')
```
