```python
import pandas as pd
import numpy 

#Read in raw data file into pandas df
data = pd.read_csv(r'C:\Users\Jordan\Documents\CodeKY\Capstone_Project_Repo\Data\famous-birthdates.txt', delimiter = " ")
data.head(10)
```

```python
#number of rows before any cleaning
len(data)
```

```python
#drop rows without a birthDate
data.dropna(subset = ['birthDate'], inplace=True)
len(data)
```

```python
#convert birthdate to date/time data type, 
data['birthDate'] = pd.to_datetime(data['birthDate'], errors='coerce')
#then convert to a day of the year
data['Date_Id'] = data['birthDate'].dt.dayofyear
#drop columns where the date_id is NULL
data.dropna(subset = ['Date_Id'], inplace=True)
data.head(20)
```

```python
#fixing instances where the person goes by a single name 
# Copy 'LastName' to 'FirstName' where 'FirstName' is null
mask = data['firstname'].isnull()
data.loc[mask, 'firstname'] = data.loc[mask, 'lastname']

# Set 'LastName' to null for the rows where 'FirstName' was null
data.loc[mask, 'lastname'] = pd.NA

# Display the DataFrame after the operation
data.head(10)
```

```python
#combine first and last name columns to create a display name
def combine_columns(row):
    return str(row['firstname']) + ' ' + row['lastname']
data['DisplayName'] = data.apply(combine_columns, axis=1)
data.head(30)
```

```python
data = data.drop(columns=['articleNum','birthDate', 'birthMonth', 'birthDay', 'zodiac'])
data = data.rename(columns={'index': 'PersonID', 'name': 'ImportName', 'lastname': 'LastName','firstname': 'FirstName', 'Date_Id': 'DateID'})
data.head(30)
```

```python
#fix instances where single names did not carry over
data['DisplayName'].fillna(data['FirstName'], inplace=True)
data.head(40)
```

```python
#Reset Index to create a Person_ID
data = data.reset_index().rename(columns={"index":"Person_ID"})
data.head(40)
```

```python
filtered_rows = data[data['DisplayName'].str.contains(' of ', case=False, na=False)]

# Replace values in 'DisplayName' with values from 'ImportName'
data.loc[filtered_rows.index, 'DisplayName'] = filtered_rows['ImportName']
data.head(30)
```

```python
#write dataframe to previously created database table Birthdate_Date_Import

import sqlite3
connection = sqlite3.connect('Zodiac_Analysis.db')
data.to_sql('People', connection, if_exists='replace')
```
